# Cleaning Nulls

```python

df = df.dropna(axis = 1, thresh = 2176) #thresh parameter keeps any column with less than half the necessary data
```

This drops 10 columns from our dataframe.  Not a terrible loss!  
From here we can imput our data better.

```python
df.isna().sum()
```

I want a more thorough null-imputation method, so I'll backfill values up to a limit of 25 values, and then fill with the median otherwise.

```python
df = df.fillna(method = 'bfill', limit =25) #limit of 25 makes sure the backfill doesn't run forever
df = df.fillna(value = df.median()) #median value is less susceptible to being pulled by high outliers
```

```python
df.isna().sum()
```

# Dropping unwanted columns

I don't want the columns outlining the margins of error for each feature, nor do I need the 'controversial' flag.

```python
df = df.drop(columns = ['pl_orbpererr1','pl_orbpererr2','pl_orbperlim','pl_orbsmaxlim','pl_radeerr1','pl_radeerr2',
                  'pl_radelim','pl_radjerr1','pl_radjerr2','pl_radjlim','pl_bmasselim','pl_bmassjlim','pl_bmassprov',
                  'pl_orbeccenlim','pl_insolerr1','pl_insolerr2','pl_insollim','pl_eqtlim','ttv_flag'])
```
